Standard Scaler

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train) # it will learn from X_train how the values are and then it will store mean value and standard deviation value at scaler 
# fit mean 1.need to calculate mean for all columns & standard deviation
# while learning it will learn from train data and while transforming it will do for both tain and test data

# transform train and test sets
X_train_scaled = scaler.transform(X_train) # you will take each x_train and transform them 
X_test_scaled = scaler.transform(X_test)
```

```python
# here in the above scaling it will take as data frame but provides output as numpy then we have to convert the array into dataframe then
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```
